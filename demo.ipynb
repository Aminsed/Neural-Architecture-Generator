# Random Neural Network Architecture Generator
This notebook demonstrates how to use the `RandomArchitectureGenerator` class to generate random neural network architectures, visualize them using NetworkX and Matplotlib, and convert them to TensorFlow models.

#### Importing the necessary modules:

```python
import warnings
warnings.filterwarnings("ignore")

from architecture_generator import RandomArchitectureGenerator
```

### Example 1: Generating a random architecture with default parameters

In this example, we'll generate a random neural network architecture using the default parameters:

```python
input_shape = (10,)  # Input shape
num_classes = 3      # Number of classes
max_layers = 6       # Maximum number of hidden layers
max_neurons = 12     # Maximum number of neurons per hidden layer
skip_connection_prob = 0.3  # Probability of adding skip connections

generator = RandomArchitectureGenerator(input_shape, num_classes, max_layers, max_neurons, skip_connection_prob)
G = generator.create_random_architecture()
```

The `create_random_architecture` method generates a random neural network architecture based on the provided parameters. It returns a NetworkX directed graph (`G`) representing the architecture.

The generated architecture will be visualized using NetworkX and Matplotlib. Each node in the graph represents a neuron, and each edge represents a connection between neurons. The nodes are color-coded based on their layer (input, hidden, or output).

### Example 2: Converting the generated architecture to a TensorFlow model

After generating the random architecture, we can convert it to a TensorFlow model using the `convert_to_tensorflow` method.


```python
model = generator.convert_to_tensorflow(G)
model.summary()
```

The `convert_to_tensorflow` method takes the generated graph (`G`) and converts it to a TensorFlow model. The resulting model will have the same structure as the generated architecture, with the appropriate number of layers and neurons.

The `model.summary()` statement prints a summary of the converted TensorFlow model, showing the layer types, output shapes, and number of parameters.

### Example 3: Customizing the architecture parameters

We can customize the parameters of the `RandomArchitectureGenerator` class to generate architectures with different characteristics.


```python
input_shape = (8, 8, 1)
num_classes = 5
max_layers = 8
max_neurons = 15
skip_connection_prob = 0.5

generator = RandomArchitectureGenerator(input_shape, num_classes, max_layers, max_neurons, skip_connection_prob)
G = generator.create_random_architecture()
model = generator.convert_to_tensorflow(G)
model.summary()
```

In this example, we've customized the `input_shape` to include the channel dimension (1 for grayscale images), reduced the `max_layers` to 3, set the `max_neurons` to 64, and increased the `skip_connection_prob` to 0.5.

The generated architecture will have a different structure based on these modified parameters. The converted TensorFlow model will reflect these changes accordingly.
